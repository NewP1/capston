```python
import os
from glob import glob

file_path = r'C:\Users\jaehy\Desktop\darknet\data\images/'
basename = r'C:\Users\jaehy\Desktop\darknet\data\list/normal_train.txt'
file_list = os.listdir(file_path)
print(file_list)
```

```python
f = open(basename, 'r')
lines = f.readlines()
f.close()
cnt = 0
f = open(r'C:\Users\jaehy\Desktop\darknet\data\list/new_normal_train.txt', 'w')
for name in lines:
    tmp = name.rstrip()[12:-4]
#     print(tmp)
    change_list = [os.path.join(file_path, tmp + '.png'), os.path.join(file_path, tmp + '.txt')]
    print(change_list)
    for file in change_list:
        if file[-3:] == 'png':
            dst = 'frame_' + str(cnt) + '.png'
            f.write('data/images/' + dst + '\n')
        else :
            dst = 'frame_' + str(cnt) + '.txt'
        dst = os.path.join(file_path, dst)
        os.rename(file, dst)
    cnt += 1
f.close()
```
