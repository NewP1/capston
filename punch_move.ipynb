```python
from glob import glob
import shutil
import os
from tqdm import tqdm
import time


```

```python
def makedirs(path): 
   try: 
        os.makedirs(path) 
   except OSError: 
       if not os.path.isdir(path): 
           raise
```

# punch

```python
path = r'C:\Users\jaehy\Desktop\punch_labeling_608\data\images/'
des = r'C:\Users\jaehy\Desktop\new_punch/'
file_list = glob(path + '*.txt')

arr = [0] * 6
cnt = 0

print(file_list)
```

```python
for file in file_list:
    tmp = file[54:-4]
#     print(tmp)
    change_list = [os.path.join(path, tmp + '.png'), os.path.join(path, tmp + '.txt')]
    des_list = [os.path.join(des, tmp + '.png'), os.path.join(des, tmp + '.txt')]
    with open(file, 'r') as f:
        lines = f.readlines()
    for line in lines:
        index = int(list(line.split())[0])
        if index == 2:
            shutil.move(change_list[0], des_list[0])
            shutil.move(change_list[1], des_list[1])
            break
            
```
