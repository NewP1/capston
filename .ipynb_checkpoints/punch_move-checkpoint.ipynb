```python
from glob import glob
import shutil
import os

path = r'D:\4\images/'
des_punch = r'D:\4\punch/'
file_list_punch = glob(path + '*.txt')

arr_punch = [0] * 6
cnt_punch = 0

print(file_list_punch)
```

# punch

```python
for file in file_list_punch:
    tmp = file[12:-4]
    print(tmp)
    change_list = [os.path.join(path, tmp + '.png'), os.path.join(path, tmp + '.txt')]
    des_list = [os.path.join(des_punch, tmp + '.png'), os.path.join(des_punch, tmp + '.txt')]
    with open(file, 'r') as f:
        lines = f.readlines()
    for line in lines:
        index = int(list(line.split())[0])
        if index == 2:
            shutil.move(change_list[0], des_list[0])
            shutil.move(change_list[1], des_list[1])
            break
            
```
