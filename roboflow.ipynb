```python
for file in file_list:
    with open(file, 'r') as f:
        lines = f.readlines()
    with open(file, 'w') as f:
        for line in lines:
            tmp = list(line.split())
            index = int(tmp[0])
            if index == 0:
                tmp[0] = '4'
                f.write(' '.join(tmp)+'\n')
            elif index == 1:
                tmp[0] = '3'
                f.write(' '.join(tmp) + '\n')
            elif index == 2:
                tmp[0] = '2'
                f.write(' '.join(tmp) + '\n')
    
```
